```python
#Load the tweet data 
import pandas as pd
```

```python
data = pd.read_csv('twitter.csv')
data
```

### Text Processing
* Removing unwanted symbols
* Convert the text into lower case
* Remove the punctuations
* Remove stopwords

```python
data['tweet'] = data['tweet'].str.replace('\$\w+','',regex=True)
data
```

```python
data['tweet'] = data['tweet'].str.replace('\.\w+','',regex=True)
data
```

```python
#Make it into lower case
data['lower case'] = data['tweet'].str.lower()
data
```

```python
#we need the text and number from tweets, we will remove all the symbols, punctuations
data['clean text'] = data['lower case'].str.replace("[^a-z0-9]"," ",regex=True)
data
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_list = stopwords.words('english')
```

```python
stop_list
```

```python
data
```

```python
def sw(x):
    x = [word for word in x.split() if word not in stop_list]
    return " ".join(x)
```

```python
data['final'] = data['clean text'].apply(sw)
data
```

```python
!pip install WordCloud
```

```python
text_final = ' '.join(data['final'])
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(width=800,height=400,background_color='white').generate(text_final)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

### Sentiment Analysis

```python
!pip install vaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sentiment = SentimentIntensityAnalyzer()
```

```python
data_df = data[['company','tweet']]
data_df
```

```python
data_df['Sentiment Score'] = [sentiment.polarity_scores(x)['compound'] for x in data_df['tweet']]
data_df
```

```python
grouped_data = data_df.groupby('company')[['Sentiment Score']].mean()
grouped_data
```

```python
#Top 5 positive tweets
grouped_data.sort_values(by='Sentiment Score',ascending=False).head(5)
```

```python
#Bottom 5 negative tweets
grouped_data.sort_values(by='Sentiment Score').head(5)
```
